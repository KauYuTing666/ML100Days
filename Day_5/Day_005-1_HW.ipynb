```python
import pandas as pd
import numpy as np
```

```python
np.random.rand(10)
print(type(np.random.rand(10)))
```

```python
data = {'國家':['台灣','中國','韓國','日本'] ,
        '人口':(np.random.rand(4)*1000).tolist() }  #隨機產生一個有4個element的ndarray，再將此ndarray轉為list
print(data)
print(type(data))
data = pd.DataFrame(data) #用pd.DataFrame將dict做成表格
data
```

```python
max_num=data.loc[:,"人口"].max()  #取出最多人口數值
print(max_num)
print(data['人口'] == max_num)
max=data.loc[data['人口'] == max_num] #用布林函數取值
max
```
