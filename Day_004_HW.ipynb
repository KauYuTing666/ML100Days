#### 截取部分資料

panda擷取資料的方法: [ ]、loc、iloc

loc用label選取資料，label是各行、列的名稱

iloc用index選取資料，index是各行、列的編號(從0開始)


```python
app_train[0:4]  #獲取部分列資料
```

```python
app_train.iloc[0:3,[1,2,4]] #iloc用index擷取資料  #這裡的0:3是指index為0開始到小於3的列，即為第0列到第2列
```
